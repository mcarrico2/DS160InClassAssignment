### Load Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Load company sales data and store to `sales`

```python
sales=pd.read_csv("company_sales_data.csv")
sales.head()
```

## Use pandas to impute missing valies if there is any

```python
sales.info()
```

```python
sales['Shirt'].fillna(sales['Shirt'].mode()[0], inplace=True)
```

```python
sales.info()
```

## Exercise 1.Read Bags, Top, Shirt and Pant sales data and show it  using a multiline plot


```python
fig=plt.figure(figsize=(16,8))
plt.plot(sales['month_number'],sales['Bags'],marker='o',label='Bags Sales Data')
plt.plot(sales['month_number'],sales['Top'],marker='o',label='Top Sales Data')
plt.plot(sales['month_number'],sales['Shirt'],marker='o',label='Shirt Sales Data')
plt.plot(sales['month_number'],sales['Pant'],marker='o', label='Pants Sales Data')
plt.title('Sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.show()
```

## Exercise 2: Read Pant and Shirt sales data of each month and show it using a scatter plot


```python
fig=plt.figure(figsize=(8,5))
plt.scatter(sales['month_number'], sales['Shirt'], marker='o', label='Shirt Sales Data')
plt.scatter(sales['month_number'], sales['Pant'], marker='o', label='Pant Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.legend(loc='upper left')
plt.show()
```

## Exercise 3: Read Jeans and Top product sales data and show it using the bar chart

Note:
-  move x axis of jeans .25 left and Top .25 right and add width
-  plt.bar(sales['month_number']-.25, sales['Jeans'],color='red',width=.30, label="Jeans sales data")

```python
fig=plt.figure(figsize=(8,5))
plt.bar(sales['month_number']-.25, sales['Jeans'],color='red',width=.30, label="Jeans sales data")
plt.bar(sales['month_number']+.25, sales['Top'],color='blue',width=.30, label="Top sales data")
plt.title('Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Units in Number')
plt.legend(loc='upper left')
plt.show()
```

## Exercise 4: Read the total profit of each month and show it using the histogram to see the most common profit ranges

## Exercise 5: Calculate `total sale` data for last year for each product and show it using a Pie chart

- salesData=[sales['Bags'].sum(),.......]

```python
salesData=[sales['Bags'].sum(),
           sales['Tie'].sum(),
           sales['Shirt'].sum(),
          sales['Jeans'].sum(),
          sales['Top'].sum(),
          sales['Pant'].sum(),]
```

```python
names=('Bags','Tie','Shirt','Jeans','Top','Pant')
plt.pie(salesData,
       autopct='%1.1f%%',
       labels=names)
plt.title('Sales Data')
plt.show()
```

### Exercise 6: Read all product sales data and show it using the stack plot. Save the result as `stack_sales.png`


```python
x = sales['month_number']
y = np.vstack([sales['Bags'],
           sales['Tie'],
           sales['Shirt'],
          sales['Jeans'],
          sales['Top'],
          sales['Pant']])

labels=['Bags','Tie','Shirt','Jeans','Top','Pant']
plt.stackplot(x,y,labels=labels,edgecolor='black')
plt.legend(loc='upper left')
plt.show()
```
